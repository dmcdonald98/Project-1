For this project I will be looking into the opinions of Americans with data gathered by the General Social Survey on the right of free speech when it comes to several more 'extreme' viewpoints. How far and in which directions does the American public believe the first amendment should extend? To begin I gathered the data from the 11 most relevant questions in the survey. The survey asked variations of 3 different questions as it relates to certain types of 'extreme speech'. Firstly, if an extremist wanted to make a public speech, should they be allowed to? Secondly, if an extremist had a book detailing these extreme views in a library, should it be removed? And thirdly, if a teacher or professor was found to hold these extreme views, should they be removed from teaching? The 4 forms of extreme speech inquired about are as follows: racist speech, communist speech, islamic extremist speech, and anti-religion extremist speech.  This project seeks to compare and draw analysis from these inquiries and discover if there are any biases towards or against certain types of extremist speech, and additionally the if the method of communication makes a difference in the responses. The data analysis found that respondents had the most problem with racist and radical islamic speech, additionally they had the greatest problem when those forms of extreme speech were said to be in a teaching role. Americans in the survey seem to be most lenient when it comes to literature, the ratio of those who would would allow the books to remain vs getting rid of the offending books was higher than each of the other catagories in question and for each catagory of extreme speech. The data for public speeches was similar to that of the data involving whether or not an offensive book should be removed with the major difference being that more respondents said they would prefer to prevent a racist or radical Islamic public speech by a slight margin, where the opposite was true for books.  Teaching was the most controversial topic by far with those opposed to racist and radical Islamic teachers around the 40 percentile and over 30% more respondents chose to respond to these two questions than almost all of the other questions used for the project.
